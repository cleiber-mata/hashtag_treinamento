# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
import pyautogui as pg
import pyperclip as pclip
import pandas as pd
import time
pg.PAUSE = 0.5
link_drive = r"https://drive.google.com/drive/folders/1X4qQW6ODmQlWei1Ei_YZiS_vuFyH96F8"
email_link = r"https://mail.google.com/"
# pg.click -> clicar
# pg.press -> apertar 1 tecla
# pg.hotkey -> conjunto de teclas
# pg.write -> escreve um texto

# Passo 1: Entrar no sistema da empresa (no nosso caso é o link do drive)
pg.press("win")
pg.write("chrome")
time.sleep(1)
pg.press("enter")
pg.write(link_drive)
pg.press("enter")
time.sleep(3)

# Passo 2: Navegar no sistema e encontrar a base de vendas (entrar na pasta exportar)
#pg.click(, clicks=2) # Clicar na pasta Exportar

# Passo 3: Fazer o download da base de vendas
pg.click(x=365, y=480) # Selecionar o arquivo
pg.click(x=1265, y=483) # Clicar nos 3 pontinhos
time.sleep(3)
pg.click(x=992, y=381) # Clicar em fazer o download
pg.click(x=527, y=440) # Salvar como
time.sleep(5)
```

### Vamos agora enviar um e-mail pelo gmail

```python

# Passo 4: Calcular os indicadores;
# Abrir a base de dados;
caminho = r"D:\Cleiber\Documents\curso_python01\hashtag_treinamento\projeto_hashtag\Seu_Primeiro_Projeto_Python\Aula 1\Exportar\Vendas - Dez.xlsx"
tabela = pd.read_excel(caminho)
#Ver as informações da base de dados;
display(tabela)
# Somar o faturamentos;
faturamento = tabela["Valor Final"].sum()
# Somar a quantidade de produtos.
qtde_produtos = tabela["Quantidade"].sum()

print("Faturamento:", faturamento)
print("Quantidade produtos:", qtde_produtos)

# Passo 5: Enviar um e-mail para a diretoria com os indicadores.
pg.press("win")
pg.write("chrome")
time.sleep(2)
pg.press("enter")
pg.write(email_link)
pg.press("enter")
time.sleep(3)
pg.click(x=1166, y=147) # Abrir gmail.
time.sleep(3)
pg.click(x=97, y=216) # Clicar em escrever.
time.sleep(2)


# Destinatário
pg.write("cleiber.mata@gmail.com")
time.sleep(2)
# Assunto
pg.press("tab")
time.sleep(1)
pg.press("tab")
pclip.copy("Relatório Teste Projeto Hashtag")
pg.hotkey("ctrl", "v")
time.sleep(2)
pg.press("tab")
# Corpo do e-mail
texto = f'''
Prezados,

Segue o relatório de vendas.

Faturamento: R${faturamento:,.2f}
Quantidade de produtos vendidos: {qtde_produtos:,}

Qualquer dúvidas estou a disposição.
abs,
Cleiber Mata.
'''
pclip.copy(texto)
pg.hotkey("ctrl", "v")
time.sleep(2)

# Enviar e-mail.
pg.click(x=852, y=695) # Clicar em enviar.



```
