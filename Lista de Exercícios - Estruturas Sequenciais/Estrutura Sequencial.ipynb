#### 4. Fa√ßa um Programa que pe√ßa as 4 notas bimestrais de um aluno e mostre a m√©dia de todas as notas.

```python
#chatGPT
def ler_nota(mensagem):
    while True:
        valor = input(mensagem).strip()
        
        if valor == "":
            print("‚ö†Ô∏è Digite uma nota v√°lida (n√£o pode ficar em branco).")
            continue
        
        try:
            nota = float(valor)
            if 0 <= nota <= 10:
                return nota
            else:
                print("‚ö†Ô∏è A nota deve estar entre 0 e 10.")
        except ValueError:
            print("‚ö†Ô∏è Digite apenas n√∫meros. Ex: 7.5")

# Entrada das notas com valida√ß√£o
bim1 = ler_nota("Nota 1¬∫ bimestre: ")
bim2 = ler_nota("Nota 2¬∫ bimestre: ")
bim3 = ler_nota("Nota 3¬∫ bimestre: ")
bim4 = ler_nota("Nota 4¬∫ bimestre: ")

media = (bim1 + bim2 + bim3 + bim4) / 4

print("====== NOTA BIMESTRAL ======")
print(f"Nota 1¬∫ Bimestre: {bim1:.2f}")
print(f"Nota 2¬∫ Bimestre: {bim2:.2f}")
print(f"Nota 3¬∫ Bimestre: {bim3:.2f}")
print(f"Nota 4¬∫ Bimestre: {bim4:.2f}")

print("\n======= M√âDIA ANUAL =======")

if media >= 6.0:
    print(f"M√©dia anual do aluno: {media:.2f} -> Aprovado üéâ")
else:
    print(f"M√©dia anual do aluno: {media:.2f} -> Reprovado ‚ùå")
```

#### 12. Fa√ßa um Programa que pergunte quanto voc√™ ganha por hora e o n√∫mero de horas trabalhadas no m√™s.

```python
ganho_hora = float(input("Qual o valor da hora trabalhada?"))
horas_mes = float(input("Quantas horas trabalhadas por m√™s?"))
```

#####  Calcule o sal√°rio bruto (horas * salario por hora)

```python
salario_bruto = ganho_hora * horas_mes
```

##### Calcule o desconto do IR (11% do sal√°rio bruto)

```python
desconto_ir = salario_bruto * 0.11
```

##### Calcule o desconto do INSS (8% do sal√°rio bruto)

```python
desconto_inss = salario_bruto * 0.08
```

##### Calcule o desconto do sindicato (5% do sal√°rio bruto)

```python
desconto_sind = salario_bruto * 0.05
```

##### Calcule o sal√°rio l√≠quido (sal√°rio bruto - descontos)

```python
descontos = desconto_ir + desconto_inss + desconto_sind
salario_liquido = salario_bruto - descontos

print("\n======SAL√ÅRIO BRUTO=====")
print(f"Sal√°rio Bruto do Funcionario: R${salario_bruto:.2f}")
print("\n======DESCONTOS=====")
print(f"Desconto IR: R${desconto_ir:.2f}")
print(f"Desconto INSS: R${desconto_inss:.2f}")
print(f"Desconto Sindicato: R${desconto_sind:.2f}")
print("\n======SAL√ÅRIO LIQUIDO=====")
print(f"Sal√°rio Liquido: R${salario_liquido:.2f}")
```

#### 13. Fa√ßa um programa para uma loja de tintas. O programa dever√° pedir o tamanho em metros quadrados da √°rea a ser pintada. Considere que a cobertura da tinta √© de 1 litro para cada 3 metros quadrados e que a tinta √© vendida em latas de 18 litros, que custam R\$ 80,00. Informe ao usu√°rio a quantidades de latas de tinta a serem compradas e o pre√ßo total. (para simplifica√ß√£o nesse momento, n√£o se preocupe em arredondar a quantidade de latas a serem compradas - vamos trabalhar isso em breve)

```python
import math
area_pintar = float(input("Qual a √°rea a ser pintada em metros quadrados? "))

rendimento_lata_18l = 54 #m¬≤
consumo_lata = math.ceil(area_pintar / rendimento_lata_18l)

custo_lata = 80 #R$
custo_total = consumo_lata * custo_lata

print(f"\nSer√° necessario {consumo_lata} latas de tinta, por um valor de R${custo_total:.2f}")
```

#### 14. Fa√ßa um programa que pe√ßa o tamanho de um arquivo para download (em MB) e a velocidade de um link de Internet (em Mbps), calcule e informe o tempo aproximado de download do arquivo usando este link (em minutos).

Detalhe: MB significa megabyte, Mb (com b min√∫sculo) significa megabit. Um megabit √© 1/8 de um megabyte. 

```python
arquivo_megabyte = int(input("Qual o tamanho do arquivo para Downloard em MB? "))
velocidade_mbps = int(input("Qual a velocidade da sua internet em Mbps? "))
arquivo_megabit = arquivo_megabyte * 8
tempo_download = arquivo_megabit / velocidade_mbps
tempo_minutos = tempo_download / 60

print(f'O tempo de download √© de {tempo_minutos:.2f} minutos')
```
